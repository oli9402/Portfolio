# Convert Excel File into textfile

We are using following [dataset](https://naehrwertdaten.ch/de/downloads/) and MAC terminal.

Using Pyhton to convert to semicolon text

On mac: `pip install openpyxl`.

Next we convert the excel file into a new semicolon text file with help of chatgpt:

- Change directory in terminal the folder containing excel file


```{code}

python3 -c "
import csv
import openpyxl

wb = openpyxl.load_workbook('Schweizer_Nahrwertdatenbank.xlsx')
sheet = wb.active

with open('output.txt', 'w', newline='') as f:
    writer = csv.writer(f, delimiter=';')
    for row in sheet.iter_rows(values_only=True):
        writer.writerow(row)
"

```



We check the first three lines of s_nw.csv with 

- `$ head -n 3 output.txt`

Since the first two lines are not needed we delete them for further processing

- `sed -i '' '1,2d' output.txt`

With excel we explore which column we need for our dashboard: 

- 4: name of product
- 6: category of product
- 8: unit
- 9: energy (kJ)
- 12: energy, calories (kcal)
- 15: fett, total(g)
- 18: fettsäure, gesättigt(g)
- 21: fettsäure, einfach ungesättigt(g)
- 24: fettsäure, mehrfach ungesättigt(g)
- 27: cholesterin (mg)
- 30: kohlenhydrate, verfügbar(g)
- 33: zucker(g)
- 36: stärke (g)
- 39: nahrungsfasern(g)
- 42: Protein (g)
- 45: salz (g)
- 48: alkohol(g)
- 51: wasser (g)
- 54: Vitamin A-Aktivität, RE (µg-RE)
- 57: Vitamin A-Aktivität, RAE (µg-RE)
- 60: Retinol (µg)
- 63: Betacarotin-Aktivität (µg-BCE)
- 66: Betacarotin (µg)
- 69: Vitamin B1 (Thiamin) (mg)
- 72: Vitamin B2 (Riboflavin) (mg)
- 75: Vitamin B6 (Pyridoxin) (mg)
- 78: Vitamin B12 (Cobalamin) (µg)
- 81: Niacin (mg)
- 84: Folat (µg)
- 87: Pantothensäure (mg)
- 90: Vitamin C (Ascorbinsäure) (mg)
- 93: Vitamin D (Calciferol) (µg) 
- 96: Vitamin E (α-Tocopherol) (mg)
- 99: Kalium (K) (mg)
- 102: Natrium (Na) (mg)
- 105: Chlorid (Cl) (mg)
- 108: Calcium (Ca) (mg)
- 111: Magnesium (Mg) (mg)
- 114: Phosphor (P) (mg)
- 117: Eisen (Fe) (mg)
- 120: Jod (I) (µg)
- 123: Zink (Zn)  (mg)
- 126: Selen (Se) (µg)

Since we have some columns with format "...,..." (e.g, energy, calories(csv)) we had to convert the file into semicolon text rather than csv, since the `cut` command would have created problems. But with semicolon there are no problems  

`cut -d';' -f4,6,8,9,12,15,18,21,24,27,30,33,36,39,42,45,48,51,54,57,60,63,66,69,72,75,78,81,84,87,90,93,96,99,102,105,108,111,114,117,120,123,126 output.txt > s_nw.txt`
